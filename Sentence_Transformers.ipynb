```python
from sentence_transformers import SentenceTransformer
```

```python
sentences = [
  "The curious cat explored every nook and cranny of the abandoned house.",
  "The aroma of freshly baked bread filled the kitchen with warmth.",
  "Despite the rain, the children eagerly built sandcastles on the beach.",
  "Artificial intelligence is revolutionizing many aspects of our lives.",
  "The ancient ruins whispered secrets of a forgotten civilization.",
  "The astronaut gazed in awe at the breathtaking view of Earth from space.",
  "The energetic puppy chased its tail in a playful frenzy.",
  "The talented musician captivated the audience with their performance.",
  "The library offered a haven of peace and knowledge.",
  "Sustainable practices are essential for protecting our environment.",
  "The historical novel transported the reader to another time and place.",
  "The solution to the complex problem seemed elusive.",
  "The vibrant colors of the autumn leaves painted a stunning landscape.",
  "The bond between a dog and its owner is truly special.",
  "Curiosity is the driving force behind scientific discovery.",
  "The delicious meal satisfied everyone's hunger.",
  "The importance of clear communication cannot be overstated.",
  "The tranquil garden provided a sense of serenity.",
  "The innovative technology has the potential to change the world.",
  "The intricate tapestry showcased the skill of the artisan."
]

model_name = 'bert-base-nli-mean-tokens'
```

```python
model = SentenceTransformer(model_name)
```

```python
sentence_vecs = model.encode(sentences)
```

```python
sentence_vecs.shape
```

```python
sentence_vecs
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(
    [sentence_vecs[0]],
    sentence_vecs[1:]

)
```
