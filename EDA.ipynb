```python


# Columns to plot
columns_to_plot = ['product_group_name', 'department_name', 'index_name', 'index_group_name', 'section_name', 'garment_group_name']

# Plot top 12 value counts for each column
for column in columns_to_plot:
    if column in Article.columns:
        plt.figure(figsize=(10, 6))

        # Get the top 12 value counts
        top_12_values = Article[column].value_counts().nlargest(12).reset_index()
        top_12_values.columns = [column, 'count']

        sns.barplot(data=top_12_values, y=column, x='count', palette='viridis', dodge=False)
        
        plt.title(f'Top 12 Value Counts of {column}')
        plt.xlabel('Counts')
        plt.ylabel(column)
        plt.show()
    else:
        print(f"Column '{column}' does not exist in the DataFrame.")
```

```python

# Columns to plot
columns_to_plot = ['product_group_name', 'index_name', 'index_group_name', 'section_name', 'garment_group_name']

# Plot top 12 value counts for each column with department_name as hue
for column in columns_to_plot:
    plt.figure(figsize=(14, 15))
    
    # Get the top 12 value counts for the column
    top_12_values = Article[column].value_counts().nlargest().index
    filtered_df = Article[Article[column].isin(top_12_values)]
    
    sns.countplot(data=filtered_df, y=column, hue='department_name', order=top_12_values, palette='viridis')
    
    plt.title(f'Top 12 Value Counts of {column} with Department Name as Hue')
    plt.xlabel('Counts')
    plt.ylabel(column)
    plt.legend(title='Department Name', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python


# List of columns to use for concatenation
selected_columns = ['prod_name', 'product_group_name', 'department_name', 'index_name', 'index_group_name', 'section_name', 'garment_group_name','detail_desc']

# Create dfQuery DataFrame with 'product_code' and 'Query' columns
dfQuery = pd.DataFrame(columns=['product_code', 'Query'])

# Function to combine text from pairs of columns into a single string, preserving 'product_code'
def combine_columns(row, columns):
    query_parts = []
    for i in range(0, len(columns), 2):
        chunk = columns[i:i+1]
        query_parts.append(' '.join(row[col] for col in chunk))
    return ', '.join(query_parts)

# Combine text from pairs of columns into a single string for each row
dfQuery['Query'] = dfArticle.apply(lambda x: combine_columns(x, selected_columns), axis=1)
dfQuery['product_code'] = dfArticle['product_code']  # Preserve 'product_code' column

# Display the dfQuery DataFrame
dfQuery.head(20)
```

```python
import pandas as pd
from itertools import combinations

# Function to combine text from columns into a single string, preserving 'product_code'
def combine_columns(row, columns):
    words = set()
    for col in columns:
        if col == 'product_code':
            words.add(row[col])  # Preserve 'product_code' as it is
        else:
            words.update(set(row[col].split()))
    return ' '.join(words)

# List of columns to iterate over
selected_columns = ['prod_name', 'product_group_name', 'department_name', 'index_name', 'index_group_name', 'section_name', 'garment_group_name']

# Generate all combinations of 2 columns
all_column_combinations = combinations(selected_columns, 2)

# Iterate over all combinations and create 'Query' column for each row
for column_combination in all_column_combinations:
    dfArticle[column_combination] = dfArticle.apply(lambda x: combine_columns(x, list(column_combination)), axis=1)

# Create dfQuery DataFrame with 'product_code' and 'Query' columns
dfQuery = dfArticle[['product_code'] + list(all_column_combinations)]

# Display the first few rows of dfQuery
dfQuery.head()
```

```python
# Function to count words in a text
def count_words(text):
    return len(text.split())

#---------------------------------------------------------------------------------------------------
# Calculate median word count for each column
median_prod_name = Article['prod_name'].apply(count_words).median()
median_product_type_name = Article['product_type_name'].apply(count_words).median()
median_product_group_name = Article['product_group_name'].apply(count_words).median()

#---------------------------------------------------------------------------------------------------

median_section_name = Article['section_name'].apply(count_words).median()
medan_garment_group_name = Article['garment_group_name'].apply(count_words).median()
median_detail_desc = Article['detail_desc'].apply(count_words).median()

#==========================================================================================

print("Median Number of Words in 'prod_name':", median_prod_name)
print("Median Number of Words in 'product_group_name':", median_product_group_name)
print("Median Number of Words in 'product_type_name':", median_product_type_name)
#----------------------------------------------------------------------------------------------------

print("Median Number of Words in 'median_section_name':", median_section_name)
print("Median Number of Words in 'garment_group_name':", medan_garment_group_name)
print("Median Number of Words in 'median_detail_desc':", median_detail_desc)

```
